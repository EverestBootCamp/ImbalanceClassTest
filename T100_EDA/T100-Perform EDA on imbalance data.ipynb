# Frequency Plot for Class Data

```python
#Checking count for Class feature
counts = df.Class.value_counts()
zeroes = counts[0]
ones = counts[1]
zeroes_perc = (zeroes/(zeroes+ones))*100
ones_perc = (ones/(zeroes+ones))*100
print('In the class category there are {} zeroes ({:.3f}%) and {} ones ({:.3f}%).'.format(zeroes,zeroes_perc,ones,ones_perc))
```

```python
#Plotting the class category
plt.figure(figsize=(8,6))
sns.barplot(x=counts.index, y=counts)
plt.title('Count of zeroes vs. ones')
plt.ylabel('Count')
plt.xlabel('Class (0, 1)')
```

Observation :

Looks like the Class feature is unbalanced because the number of values 0 and 1 does differ.

The number of Class 0 is almost 600 times the number of Class 1.

The dataset is imbalanced so we have applied stratified sampling technique on the latter stages.

# Visualizing 3-D numeric data with Scatter Plots

```python
import random
```

```python
# random sample 200 positions for both the classes
sample_0 = random.sample(range(df[df['Class']==0].shape[0]),200)
sample_1 = random.sample(range(df[df['Class']==1].shape[0]),200)
```

```python
xs_0 = df[df['Class']==0]['T1'].reset_index(drop=True).iloc[sample_0]
ys_0 = df[df['Class']==0]['T12'].reset_index(drop=True).iloc[sample_0]
zs_0 = df[df['Class']==0]['T18'].reset_index(drop=True).iloc[sample_0]

xs_1 = df[df['Class']==1]['T1'].reset_index(drop=True).iloc[sample_1]
ys_1 = df[df['Class']==1]['T12'].reset_index(drop=True).iloc[sample_1]
zs_1 = df[df['Class']==1]['T18'].reset_index(drop=True).iloc[sample_1]
```

```python
# Visualizing 3-D numeric data with Scatter Plots
# length, breadth and depth
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xs_0, ys_0, zs_0, s=50, alpha=0.6, edgecolors='w', color='red')
ax.scatter(xs_1, ys_1, zs_1, s=50, alpha=0.6, edgecolors='w', color='blue')
ax.set_xlabel('T1', size=12)
ax.set_ylabel('T12', size=12)
ax.set_zlabel('T18', size=12)
plt.title("Distribution of 2 classes in 3D space for T1, T12 and T18 features", size=16)
ax.legend(('Class 0', 'Class 1'),loc=6)
plt.show()
```

```python
def scatter_3d(a,b,c):
    xs_0 = df[df['Class']==0][a].reset_index(drop=True).iloc[sample_0]
    ys_0 = df[df['Class']==0][b].reset_index(drop=True).iloc[sample_0]
    zs_0 = df[df['Class']==0][c].reset_index(drop=True).iloc[sample_0]

    xs_1 = df[df['Class']==1][a].reset_index(drop=True).iloc[sample_1]
    ys_1 = df[df['Class']==1][b].reset_index(drop=True).iloc[sample_1]
    zs_1 = df[df['Class']==1][c].reset_index(drop=True).iloc[sample_1]
    
    # Visualizing 3-D numeric data with Scatter Plots
    # length, breadth and depth
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')

    ax.scatter(xs_0, ys_0, zs_0, s=50, alpha=0.6, edgecolors='w', color='red')
    ax.scatter(xs_1, ys_1, zs_1, s=50, alpha=0.6, edgecolors='w', color='blue')
    ax.set_xlabel(a, size=12)
    ax.set_ylabel(b, size=12)
    ax.set_zlabel(c, size=12)
    title = "Distribution of 2 classes in 3D space for {}, {} and {} features".format(a,b,c)
    plt.title(title, size=16)
    ax.legend(('Class 0', 'Class 1'),loc=6)
    plt.show()    
```

Following 3D scatter plots are for the positively and negatively correlated features
found through the correlation part

```python
scatter_3d('T1','T3','T7')
```

```python
scatter_3d('T10','T12','T14')
```

```python
scatter_3d('T14','T16','T17')
```

```python
scatter_3d('T4','T11','T18')
```

Through these graphs one thing we can clearly notice is Class==0 are closely clustered.
But for Class==1, the data is randomly spread and seems to have outliers. Also, we can notice that the 'Class 0' & 'Class 1' are not linearly separable.
